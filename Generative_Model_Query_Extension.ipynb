```python
from transformers import pipeline

text_generator = pipeline("text-generation")

file1 = open('queries.txt', 'r')
Lines = file1.readlines()

strings= []
for i, line in enumerate(Lines):
    if line.startswith('\n'):
        continue
    elif line.startswith('ID'):
        strings.append(line)
    else:
        #print(text_generator(line, max_length=len(line.split(' '))+15, do_sample=False))
        strings.append(text_generator(line.strip('\n'), max_length=len(line.split(' '))+15, do_sample=False)[0]['generated_text'])
        for i in range(4):
            strings.append(text_generator(line.strip('\n'), max_length=len(line.split(' '))+15, do_sample=True)[0]['generated_text'])
strings[:20]

f = open("queries_extended.txt", "w")
for i in range(len(strings)):
    f.write(strings[i].strip('\n'))
    f.write('\n')
```

```python
f = open("queries_extended.txt", "w")
for i in range(len(strings)):
    f.write(strings[i].replace('\n', ' '))
    f.write('\n')
```
